```python
import numpy as np
from ripser import ripser
from persim import plot_diagrams
import matplotlib.pyplot as plt
import os

# Noisy circle
theta = np.linspace(0, 2*np.pi, 100)
X = np.column_stack([np.cos(theta), np.sin(theta)])
X += np.random.normal(0, 0.1, X.shape)

# Plot Point Cloud
plt.figure(figsize=(5,5))
plt.scatter(X[:,0], X[:,1], s=10)
plt.axis('equal')
plt.title('Noisy Circle Point Cloud')
plt.show()

# Compute + plot
dgms = ripser(X)['dgms']
plot_diagrams(dgms, show=True)
```

```python
fig, ax = plt.subplots(figsize=(8,4))
plot_diagrams(dgms, plot_only=[1], ax=ax) # H1 only

# Manual Barcode
h1=dgms[1]
fig, ax = plt.subplots(figsize=(8,4))
for i, (b, d) in enumerate(h1):
    ax.plot([b,d], [i,i], 'b-', linewidth =2)
ax.set_xlabel('Filtration value')
ax.set_title('H1 Barcode')
plt.show()
```

```python
def plot_barcodes(dgms, overlay=False):
    if overlay:
        fig, ax = plt.subplots(figsize=(8, 4))
        offset = 0
        colors = ['steelblue', 'tomato', 'seagreen']
        for dim, bars in enumerate(dgms):
            finite = bars[bars[:, 1] < np.inf]
            for i, (b, d) in enumerate(finite):
                ax.plot([b, d], [offset + i, offset + i],
                        linewidth=2, color=colors[dim % len(colors)],
                        label=f'H{dim}' if i == 0 else '')
            offset += len(finite) + 1  # gap between dimensions
        ax.set_title('All Barcodes Overlaid')
        ax.set_xlabel('Filtration value')
        ax.set_yticks([])
        ax.legend()
    else:
        fig, axes = plt.subplots(1, len(dgms), figsize=(5 * len(dgms), 4))
        colors = ['steelblue', 'tomato', 'seagreen']
        for dim, (ax, bars) in enumerate(zip(axes, dgms)):
            finite = bars[bars[:, 1] < np.inf]
            for i, (b, d) in enumerate(finite):
                ax.plot([b, d], [i, i], linewidth=2, color=colors[dim % len(colors)])
            ax.set_title(f'H{dim} Barcode')
            ax.set_xlabel('Filtration value')
            ax.set_yticks([])
        plt.tight_layout()

    plt.show()

# Usage:
plot_barcodes(dgms, overlay=False)  # side by side
plot_barcodes(dgms, overlay=True)   # stacked in one plot
```

# Threshold Comparison Plotting
## Pass thresholds as either a list ([unfiltered threshold, filtered threshold]) OR singular values
Passing as a list will create a comparison row displaying filtered vs unfiltered results
## Use Overlay to add H0 features to BC & PL
## Use Grid to compare H1 vs H0,H1 visuals (only works if `overlay=True`)

```python
def compute_circle(n_points=100, noise=0.1, max_dim=1):
    theta = np.linspace(0, 2*np.pi, n_points)
    X = np.column_stack([np.cos(theta), np.sin(theta)])
    X += np.random.normal(0, noise, X.shape)
    dgms = ripser(X, maxdim=max_dim)['dgms']
    return X, dgms
    
def plot_experiment_pair(X, dgms, noise, thresholds, overlay=False, grid=False,
                         show_pd=False, show_bc=False, show_pl=False):
    from persim import PersLandscapeApprox
    from matplotlib.gridspec import GridSpec, GridSpecFromSubplotSpec

    # normalise threshold input
    if not isinstance(thresholds, (list, tuple)):
        thresholds = [thresholds]
    single = len(thresholds) == 1

    def filter_dgm(bars, threshold):
        finite = bars[bars[:,1] < np.inf]
        return finite[(finite[:,1] - finite[:,0]) > threshold]

    def draw_barcode(ax, filtered, dims=(1,), title=None):
        colors_bc = ['tomato', 'steelblue']
        offset = 0
        for dim in dims:
            bars = filtered[dim]
            finite = bars[bars[:,1] < np.inf] if dim == 0 else bars
            for i, (b, d) in enumerate(finite):
                ax.plot([b, d], [offset+i, offset+i], linewidth=1.5,
                        color=colors_bc[dim], label=f'H{dim}' if i == 0 else '')
            offset += len(finite) + 1
        ax.set_yticks([])
        ax.tick_params(labelsize=6)
        if len(dims) > 1:
            ax.legend(fontsize=6, loc='upper right')
        if title:
            ax.set_title(title, fontsize=8)

    def draw_landscape(ax, filtered, dims=(1,), title=None):
        land_colors = {0: ['tomato', 'lightsalmon', '#f5c0a0'],
                       1: ['steelblue', 'cornflowerblue', 'lightsteelblue']}
        padded = [np.vstack([f, [[0,0]]]) if len(f) == 0 else f for f in filtered]
        for dim in dims:
            land = PersLandscapeApprox(dgms=padded, hom_deg=dim)
            land.compute_landscape()
            xs = np.linspace(land.start, land.stop, land.num_steps)
            for i, ys in enumerate(land.values[:3]):
                ax.plot(xs, ys, color=land_colors[dim][i],
                        alpha=1 - i*0.25,
                        label=f'H{dim} λ{i+1}' if i == 0 else '')
        ax.tick_params(labelsize=6)
        if len(dims) > 1:
            ax.legend(fontsize=6, loc='upper right')
        if title:
            ax.set_title(title, fontsize=8)

    n_points = len(X)
    lim = max(dgms[0][:,1][dgms[0][:,1] < np.inf].max(),
              dgms[1][:,1][dgms[1][:,1] < np.inf].max()) * 1.05

    filtered_sets = [[filter_dgm(bars, t) for bars in dgms] for t in thresholds]

    # If none explicitly requested, show all three.
    # If any are True, show only those that are True.
    any_requested = show_pd or show_bc or show_pl
    diagram_cols = []  # col_type: 'pd', 'bc', 'pl'
    if show_pd or not any_requested:
        diagram_cols.append('pd')
    if show_bc or not any_requested:
        diagram_cols.append('bc')
    if show_pl or not any_requested:
        diagram_cols.append('pl')

    n_diag_cols = len(diagram_cols)
    total_cols = 1 + n_diag_cols  # point cloud + diagram cols

    # Width ratios: point cloud slightly wider, rest equal
    cloud_ratio = max(1.0, 4 / total_cols)  # scale down if fewer cols
    width_ratios = [cloud_ratio] + [1] * n_diag_cols
    fig_width = 6 + 4 * n_diag_cols  # scale figure width with columns

    if single:
        fig = plt.figure(figsize=(fig_width, 4))
        gs = GridSpec(1, total_cols, figure=fig,
                      width_ratios=width_ratios,
                      hspace=0.08, wspace=0.4,
                      left=0.05, right=0.97, top=0.88, bottom=0.12)
    else:
        fig = plt.figure(figsize=(fig_width, 6))
        gs = GridSpec(2, total_cols, figure=fig,
                      width_ratios=width_ratios,
                      hspace=0.08, wspace=0.4,
                      left=0.05, right=0.97, top=0.88, bottom=0.1)

    fig.suptitle(f'n={n_points}  |  noise={noise}', fontsize=12,
                 fontweight='bold', x=0.05, ha='left')

    # ── point cloud (always col 0)
    ax_cloud = fig.add_subplot(gs[:, 0])
    ax_cloud.scatter(X[:,0], X[:,1], s=8, color='steelblue', alpha=0.7)
    ax_cloud.set_aspect('equal')
    ax_cloud.set_title('Point cloud', fontsize=10)
    ax_cloud.tick_params(labelsize=8)

    # ── diagram columns
    colors = ['tomato', 'steelblue', 'seagreen']

    for col_offset, col_type in enumerate(diagram_cols):
        col_idx = 1 + col_offset
        draw_fn = draw_barcode if col_type == 'bc' else draw_landscape

        if col_type == 'pd':
            # Persistence diagram — its own draw logic
            if single:
                ax_pd = fig.add_subplot(gs[0, col_idx])
                filtered = filtered_sets[0]
                for dim, bars in enumerate(filtered):
                    if len(bars):
                        ax_pd.scatter(bars[:,0], bars[:,1], s=15,
                                      color=colors[dim], label=f'H{dim}', zorder=3)
                ax_pd.plot([0, lim], [0, lim], 'k--', linewidth=0.5, zorder=1)
                ax_pd.set_xlim(0, lim); ax_pd.set_ylim(0, lim)
                ax_pd.set_xlabel('birth', fontsize=8); ax_pd.set_ylabel('death', fontsize=8)
                ax_pd.tick_params(labelsize=7)
                ax_pd.legend(fontsize=7, loc='lower right')
                ax_pd.set_title('Persistence diagram', fontsize=10)
            else:
                for row, (thresh, filtered) in enumerate(zip(thresholds, filtered_sets)):
                    ax_pd = fig.add_subplot(gs[row, col_idx])
                    for dim, bars in enumerate(filtered):
                        if len(bars):
                            ax_pd.scatter(bars[:,0], bars[:,1], s=15,
                                          color=colors[dim], label=f'H{dim}', zorder=3)
                    ax_pd.plot([0, lim], [0, lim], 'k--', linewidth=0.5, zorder=1)
                    ax_pd.set_xlim(0, lim); ax_pd.set_ylim(0, lim)
                    ax_pd.set_xlabel('birth', fontsize=8); ax_pd.set_ylabel('death', fontsize=8)
                    ax_pd.tick_params(labelsize=7)
                    ax_pd.legend(fontsize=7, loc='lower right')
                    # compute label x position proportionally
                    label_x = (col_idx / total_cols) - 0.02
                    fig.text(label_x, 0.72 - row*0.47, f'thresh={thresh}',
                             fontsize=7, color='gray', va='center',
                             rotation=90, fontfamily='monospace')
                    if row == 0:
                        ax_pd.set_title('Persistence diagram', fontsize=10)

                fig.add_artist(plt.Line2D(
                    [(1 / total_cols), 0.97], [0.5, 0.5],
                    transform=fig.transFigure,
                    color='lightgray', linewidth=0.8, linestyle='--'))

        else:
            # Barcode or Landscape
            if single and grid:
                gs_inner = GridSpecFromSubplotSpec(
                    2, 1, subplot_spec=gs[:, col_idx], hspace=0.55)
                draw_fn(fig.add_subplot(gs_inner[0, 0]), filtered_sets[0],
                        dims=(1,),   title='H₁')
                draw_fn(fig.add_subplot(gs_inner[1, 0]), filtered_sets[0],
                        dims=(0, 1), title='H₀+H₁')

            elif single and overlay:
                ax = fig.add_subplot(gs[:, col_idx])
                draw_fn(ax, filtered_sets[0], dims=(0, 1),
                        title=f'H₀+H₁  thresh={thresholds[0]}')

            elif single:
                ax = fig.add_subplot(gs[:, col_idx])
                draw_fn(ax, filtered_sets[0], dims=(1,),
                        title=f'H₁  thresh={thresholds[0]}')

            elif grid:
                gs_inner = GridSpecFromSubplotSpec(
                    2, 2, subplot_spec=gs[:, col_idx],
                    hspace=0.55, wspace=0.25)
                titles = [
                    ['H₁ unfiltered', 'H₁ filtered'],
                    ['H₀+H₁ unfiltered', 'H₀+H₁ filtered'],
                ]
                dim_sets = [(1,), (0, 1)]
                for r in range(2):
                    for c in range(2):
                        ax = fig.add_subplot(gs_inner[r, c])
                        draw_fn(ax, filtered_sets[min(c, len(filtered_sets)-1)],
                                dims=dim_sets[r], title=titles[r][c])

            elif overlay:
                gs_inner = GridSpecFromSubplotSpec(
                    2, 1, subplot_spec=gs[:, col_idx], hspace=0.55)
                for row, (thresh, filtered) in enumerate(zip(thresholds, filtered_sets)):
                    ax = fig.add_subplot(gs_inner[row, 0])
                    label = 'unfiltered' if row == 0 else f'thresh={thresh}'
                    draw_fn(ax, filtered, dims=(0, 1), title=f'H₀+H₁ {label}')

            else:
                gs_inner = GridSpecFromSubplotSpec(
                    2, 1, subplot_spec=gs[:, col_idx], hspace=0.55)
                for row, (thresh, filtered) in enumerate(zip(thresholds, filtered_sets)):
                    ax = fig.add_subplot(gs_inner[row, 0])
                    label = 'unfiltered' if row == 0 else f'thresh={thresh}'
                    draw_fn(ax, filtered, dims=(1,), title=f'H₁ {label}')

    os.makedirs('tda_outputs', exist_ok=True)

    # Build filename suffix from what's shown
    col_tags = {'pd': 'pd', 'bc': 'bc', 'pl': 'pl'}
    shown = '+'.join(col_tags[c] for c in diagram_cols)
    thresh_tag = f'th{thresholds[0]}' if single else f'th{thresholds[0]}-{thresholds[1]}'
    ovl_tag = '_ovl' if overlay else ''
    grd_tag = '_grid' if grid else ''
    filename = f'tda_outputs/circle_n{n_points}_ns{noise}_{shown}_{thresh_tag}{ovl_tag}{grd_tag}.png'

    plt.savefig(filename, dpi=150, bbox_inches='tight')
    plt.show()
```

```python
# configs: (n_points, noise, [unfiltered_thresh, filtered_thresh])
configs = [
    (100, 0.0,  [0.0, 0.0]),    # no noise — show same twice for layout consistency
    (100, 0.05, [0.0, 0.05]),
    (100, 0.15, [0.0, 0.05]),
    (100, 0.3,  [0.0, 0.1]),
    (100, 0.5,  [0.0, 0.15]),
]

single_configs = [        # no unfiltered threshold
    (50, 0.00,  0.0),    
    (50, 0.05, 0.0),
    (50, 0.15, 0.0),
    (50, 0.3, 0.0),
    (50, 0.5, 0.0),
]

for n, noise, thresholds in single_configs:
    X, dgms = compute_circle(n_points=n, noise=noise)
    plot_experiment_pair(X, dgms, noise=noise, thresholds=thresholds, overlay=False, grid=False)
    #plot_experiment_pair(X, dgms, noise=noise, thresholds=thresholds, overlay=True, grid=False, show_bc=True)
    #plot_experiment_pair(X, dgms, noise=noise, thresholds=thresholds, overlay=True, grid=True)
```
